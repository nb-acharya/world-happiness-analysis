```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12,6)

print("Libraries loaded successfully")
```

```python
#load data of all five years

df_2015 = pd.read_csv("csv/2015.csv")
df_2016 = pd.read_csv("csv/2016.csv")
df_2017 = pd.read_csv("csv/2017.csv")
df_2018 = pd.read_csv("csv/2018.csv")
df_2019 = pd.read_csv("csv/2019.csv")

print(df_2015.shape)
print(df_2016.shape)
print(df_2017.shape)
print(df_2018.shape)
print(df_2019.shape)

print("2015 columns:", df_2015.columns.tolist())
print("2016 columns:", df_2016.columns.tolist())
print("2017 columns:", df_2017.columns.tolist())
print("2018 columns:", df_2018.columns.tolist())
print("2019 columns:", df_2019.columns.tolist())
```

```python
#standardize all dataframes to have same column names

#2015
df_2015 = df_2015[['Country', 'Region', 'Happiness Rank', 'Happiness Score',
                    'Economy (GDP per Capita)', 'Health (Life Expectancy)',
                    'Freedom', 'Generosity', 'Trust (Government Corruption)']]
df_2015.columns = ['Country', 'Region', 'Rank', 'Score', 'GDP', 'Health', 'Freedom', 'Generosity', 'Corruption']
df_2015['Year'] = 2015

# 2016
df_2016 = df_2016[['Country', 'Region', 'Happiness Rank', 'Happiness Score',
                    'Economy (GDP per Capita)', 'Health (Life Expectancy)',
                    'Freedom', 'Generosity', 'Trust (Government Corruption)']]
df_2016.columns = ['Country', 'Region', 'Rank', 'Score', 'GDP', 'Health', 'Freedom', 'Generosity', 'Corruption']
df_2016['Year'] = 2016

# 2017 (dots in column names)
df_2017 = df_2017[['Country', 'Happiness.Rank', 'Happiness.Score',
                    'Economy..GDP.per.Capita.', 'Health..Life.Expectancy.',
                    'Freedom', 'Generosity', 'Trust..Government.Corruption.']]
df_2017.columns = ['Country', 'Rank', 'Score', 'GDP', 'Health', 'Freedom', 'Generosity', 'Corruption']
df_2017['Year'] = 2017

# 2018
df_2018 = df_2018[['Country or region', 'Overall rank', 'Score', 'GDP per capita',
                    'Healthy life expectancy', 'Freedom to make life choices',
                    'Generosity', 'Perceptions of corruption']]
df_2018.columns = ['Country', 'Rank', 'Score', 'GDP', 'Health', 'Freedom', 'Generosity', 'Corruption']
df_2018['Year'] = 2018

# 2019
df_2019 = df_2019[['Country or region', 'Overall rank', 'Score', 'GDP per capita',
                    'Healthy life expectancy', 'Freedom to make life choices',
                    'Generosity', 'Perceptions of corruption']]
df_2019.columns = ['Country', 'Rank', 'Score', 'GDP', 'Health', 'Freedom', 'Generosity', 'Corruption']
df_2019['Year'] = 2019

print("Cleaning done!")
```

```python
#merge all the years in one dataframe
df = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019], ignore_index=True)

print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Check missing values and data types
print("Missing values:\n", df.isnull().sum())
print("\nData types:\n", df.dtypes)
```

```python
# Fix 1: Fill missing Region
region_map = df_2015.set_index('Country')['Region'].to_dict()
df['Region'] = df['Region'].fillna(df['Country'].map(region_map))

# Fix 2: Fill missing Corruption with median
df['Corruption'] = df['Corruption'].fillna(df['Corruption'].median())

print("Missing values after fix:\n", df.isnull().sum())
```
